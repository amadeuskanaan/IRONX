```python
def return_expression_df(geneset):
    ahba_dir     = '/Users/kanaan/Google Drive/TS-EUROTRAIN/RESULTS_QSMv3/SEPT10/AHBA/'
    drop_strings = ['coords_native', 'donor_names', 'struct_id', 
                'struct_name', 'top_struct', 'Mean', 'Median', 
                'PC1', 'PC2','PC3', #'SVD1g', 'SVD2g','SVD3g', 'SVD3p', 'SVD2p','SVD3p', 
                'corrected_mni_x', 'corrected_mni_y', 'corrected_mni_z']
    df = pd.read_csv(os.path.join(ahba_dir, 'AHBA_%s.csv'%geneset),index_col=0)
    gs = [i for i in df.columns if i not in drop_strings]
    return df, gs 

IRON_H,  GS_IRON_H   = return_expression_df('IRON_HOMEOSTASIS_PCA')
IRON_D,  GS_IRON_D   = return_expression_df('IRON_D_PCA')
IRON_T2, GS_IRON_T2  = return_expression_df('IRON_TRANSPORT2_PCA')
FERRITIN,GS_FERRITIN = return_expression_df('FERRITIN_PCA')
IRON_IM, GS_IRON_IM  = return_expression_df('IRON_ION_IMPORT_PCA')
IRON_T1, GS_IRON_T1  = return_expression_df('IRON_TRANSPORT1_PCA')
IRON_R,  GS_IRON_R   = return_expression_df('IRON_RESPONSE_PCA')

genesets = {'IRON_H'  : GS_IRON_H,
            'IRON_D'  : GS_IRON_D,
            'IRON_T2' : GS_IRON_T2,
            'FERRITIN': GS_FERRITIN,
            #'IRON_T1' : GS_IRON_T1,
            'IRON_IM' : GS_IRON_IM,
            'IRON_R'  : GS_IRON_R,
           }
iron_genes= GS_IRON_H+ GS_IRON_D +GS_IRON_T2 + GS_IRON_R

def plot_str3_ahba(str_mask, cmap ):
    
    if str_mask == 'motor':
        wells = df_motor.index
    elif str_mask == 'limbic':
        wells = df_limbic.index
    elif str_mask == 'exec': 
        wells = df_exec.index
    
    
    # plot AHBA whole brain data for some genes
    drop_cols = [i for i in AHBA.columns if i not in iron_genes if i in AHBA.columns]
    df = AHBA.drop(drop_cols, axis =1)
    df = df.drop([i for i in df.index if i not in wells], axis = 0 )
    
    #scale data
    x = df.values #returns a numpy array
    min_max_scaler = preprocessing.MinMaxScaler()
    x_scaled = min_max_scaler.fit_transform(x)
    df = pd.DataFrame(x_scaled)

    #df["top_struct"] = df["top_struct"] / 10
    fig = plt.figure(figsize=(5, 10))
    sns.heatmap(df, cmap = cmap, xticklabels=False, yticklabels=False, cbar=False)
    fig.savefig(os.path.join(save_fig_dir, 'plot_ahba_%s.png'%str_mask), bbox_inches = 'tight', transparent = True)

plot_str3_ahba('motor', 'Blues')
plot_str3_ahba('limbic', 'Greens')
plot_str3_ahba('exec', 'Reds')


```

```python
def make_voxels(df, mask_name, mask_resolution):
    #https://mail.scipy.org/pipermail/nipy-devel/2012-October/008200.html
    import nibabel as nb 
    import nipy.labs.spatial_models.mroi as mroi
    from nipy.labs.spatial_models.discrete_domain import grid_domain_from_image
    
    mask_path  = '/usr/local/fsl/data/standard/MNI152_T1_%s.nii.gz' %mask_resolution
    mask = nb.load(mask_path)
    mask_array, affine = mask.get_data() > 0, mask.get_affine()
    
    try:
        os.mkdir(os.path.join(ahba_dir, 'tmp_%s'%mask_name))
    except:
        pass

    os.chdir(os.path.join(ahba_dir, 'tmp_%s'%mask_name))
    
    print '%s mask has %s wells'  %(mask_name, len(df.index))
    
    for well in df.index:
        if not os.path.isfile('%s_%s_%s.nii.gz' %(well, mask_name, mask_resolution)):
            print 'Creating sphere for well #', well
            X = IRON_H.loc[well]['corrected_mni_x']
            Y = IRON_H.loc[well]['corrected_mni_y']
            Z = IRON_H.loc[well]['corrected_mni_z']

            print X, Y , Z 
            positions = np.array([[X,Y,Z ]])
            
            # in mm (here in the MNI space)
            radii = np.array([0.5])

            domain = grid_domain_from_image(mask)
            my_roi = mroi.subdomain_from_balls(domain, positions, radii)

            # to save an image of the ROIs
            nb.save(my_roi.to_image(), '%s_%s_%s.nii.gz' %(well, mask_name, mask_resolution))     
        
    
    roi_list = ' '.join(['%s -add'% well for well in os.listdir('./') if mask_resolution in well])[0:-5]
    #print roi_list
    os.system('fslmaths %s -bin ../spheres_%s_%s' % (roi_list, mask_name, mask_resolution))
```

# combine FSL-FIRST VTK 

```python
def combine_vtk():
    from vtk import * 
    import nibabel as nb 
    reader = vtkPolyDataReader()
    append = vtkAppendPolyData()

    puta_l = '/Users/kanaan/SCR/Github/GluIRON/atlases/FIRST/MNI-L_Puta_first.vtk'
    puta_r = '/Users/kanaan/SCR/Github/GluIRON/atlases/FIRST/MNI-R_Puta_first.vtk'
    caud_l = '/Users/kanaan/SCR/Github/GluIRON/atlases/FIRST/MNI-L_Caud_first.vtk'
    caud_r = '/Users/kanaan/SCR/Github/GluIRON/atlases/FIRST/MNI-R_Caud_first.vtk'
    accu_l = '/Users/kanaan/SCR/Github/GluIRON/atlases/FIRST/MNI-L_Accu_first.vtk'
    accu_r = '/Users/kanaan/SCR/Github/GluIRON/atlases/FIRST/MNI-R_Accu_first.vtk'

    filenames = [puta_l, puta_r, caud_l, caud_r, accu_l, accu_r]
    for file in filenames:
        reader.SetFileName(file)
        reader.Update()
        polydata = vtkPolyData()
        polydata.ShallowCopy(reader.GetOutput())
        append.AddInputData(polydata)

    append.Update()    

    writer = vtkPolyDataWriter()
    writer.SetFileName('/Users/kanaan/SCR/Github/GluIRON/atlases/FIRST/STR.vtk')
    writer.SetInputConnection(append.GetOutputPort())

    writer.Write()

combine_vtk()
```
