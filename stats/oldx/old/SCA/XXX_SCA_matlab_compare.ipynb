```matlab
surf_sphere = SurfStatReadSurf({[fsaverage5_dir 'lh.sphere'],[fsaverage5_dir 'rh.sphere']});
surf_curv   = SurfStatReadData({[fsaverage5_dir 'fsaverage_curv_lh.asc'],[fsaverage5_dir 'fsaverage_curv_rh.asc']});
surf_infl   = SurfStatReadSurf({[fsaverage5_dir 'lh.inflated'],[fsaverage5_dir 'rh.inflated']});
surf_white  = SurfStatReadSurf({[fsaverage5_dir 'lh.white'],[fsaverage5_dir 'rh.white']});
surf_pial   = SurfStatReadSurf({[fsaverage5_dir 'lh.pial'],[fsaverage5_dir 'rh.pial']});

surf_mid_ct.tri   = surf_white.tri;
surf_mid_ct.coord = (surf_white.coord + surf_white.coord)./2;
```

```matlab

phenotypic_fid  = fopen(phenotypic);
phenotypic_data = textscan(phenotypic_fid,'%s%f%s%s%f','Delimiter',',','headerLines',1,'CollectOutput',1);
fclose(phenotypic_fid);


% subject 	Age 	Group 	Sex 	Site
phenotypic_id         = phenotypic_data{1};
phenotypic_age        = phenotypic_data{2};
phenotypic_group      = phenotypic_data{3}(:,1);
phenotypic_sex        = phenotypic_data{3}(:,2);
phenotypic_fd         = phenotypic_data{4};

```

```matlab
MOTOR  = zeros(length(phenotypic_id),size(surf_white.coord,2));
LIMBIC  = zeros(length(phenotypic_id),size(surf_white.coord,2));
EXEC  = zeros(length(phenotypic_id),size(surf_white.coord,2));
```

```matlab
strM_lh = strcat(features_dir, 'gsr/STR3_MOTOR/',  phenotypic_id, '_sca_z_fsaverage5_fwhm10_lh.mgh');
strM_rh = strcat(features_dir, 'gsr/STR3_MOTOR/',  phenotypic_id, '_sca_z_fsaverage5_fwhm10_rh.mgh');
```

```matlab
for i = 1:length(phenotypic_id)
    try
       lh      = SurfStatReadData1(strM_lh{i}) ;
       rh      = SurfStatReadData1(strM_rh{i}) ;
       MOTOR(i,:)  = [lh, rh];
    end
end
```

```matlab


strL_lh = strcat(features_dir, 'gsr/STR3_LIMBIC/', phenotypic_id, '_sca_z_fsaverage5_fwhm10_lh.mgh');
strL_rh = strcat(features_dir, 'gsr/STR3_LIMBIC/', phenotypic_id, '_sca_z_fsaverage5_fwhm10_rh.mgh');

for i = 1:length(phenotypic_id)
    try
       lh      = SurfStatReadData1(strL_lh{i}) ;
       rh      = SurfStatReadData1(strL_rh{i}) ;
       LIMBIC(i,:)  = [lh, rh];
    end
end
```

```matlab

strE_lh = strcat(features_dir, dntype, '/STR3_EXEC/',   phenotypic_id, '_sca_z_fsaverage5_fwhm10_lh.mgh');
strE_rh = strcat(features_dir, dntype, '/STR3_EXEC/',   phenotypic_id, '_sca_z_fsaverage5_fwhm10_rh.mgh');

for i = 1:length(phenotypic_id)
    try
       lh      = SurfStatReadData1(strE_lh{i}) ;
       rh      = SurfStatReadData1(strE_rh{i}) ;
       EXEC(i,:)  = [lh, rh];
    end
end

features = horzcat(MOTOR, LIMBIC, EXEC);
```

```matlab
fig=figure, imagesc(features,[-1,1]);
colormap('parula');
```

```matlab
% visualize mean CT of tourettome subject 
SurfStatViewData(mean(MOTOR,1), surf_mid_ct, 'SCA MOTOR Mean');
%SurfStatColLim([-1 1]);
```

```matlab
% visualize mean CT of tourettome subject 
SurfStatViewData(mean(LIMBIC,1), surf_mid_ct, 'SCA MOTOR Mean');
%SurfStatColLim([-1 1]);

```

```matlab
% visualize mean CT of tourettome subject 
SurfStatViewData(mean(EXEC,1), surf_mid_ct, 'SCA MOTOR Mean');
%SurfStatColLim([-1 1]);
```

```matlab
% code variables of interest as terms 
term_age    = term(phenotypic_age);
term_group  = term(phenotypic_group);
term_sex    = term(phenotypic_sex);
term_fd     = term(phenotypic_fd);



% buid linear model controlled for site, group, age, sex
df_model  = 1 + term_age + term_sex + term_group +term_fd;

% Estimaste model parameters
stat_model      = SurfStatLinMod(MOTOR,df_model,surf_white);

% Specify model contrast
stat_model      = SurfStatT(stat_model,term_group.Controls - term_group.Patients); 
```

```matlab
df_model_mat = df_model(:, :);
imagesc(df_model_mat, [0 63])
cmap = jet;
cmap(2,:) =[0.9 0.9 0.9];
cmap(1,:) =[0.1 .1 .1];
colormap(cmap)
```

```matlab
df_model_mat
```

```matlab
SurfStatViewData(stat_model.t,surf_mid_ct,'t-stat HC>GTS');
SurfStatColLim([-2 2]) ;
colormap(jet)
```

```matlab
% alternative 1: none - we show uncorrected p-values, good for exploration
p = 1-tcdf(stat_model.t,stat_model.df); 
SurfStatViewData(p,surf_mid_ct,'p-val uncorrected -  HC>GTS');
SurfStatColLim([0 0.05]) ;
```

```matlab
mask  = mean(transpose(ct),1) > 0.4;

clusp = 0.05; 

[pval, peak, clus, clusid] = SurfStatP(stat_model, mask, clusp);
%pval = SurfStatP(stat_model,mask); 
```

```matlab
qval = SurfStatQ(stat_model, mask);
SurfStatViewData(qval.Q,surf_mid_ct,'p-val FDR -  HC>GTS');
%colormap([0.8 .8 .8; ice/255])
%SurfStatColLim([0 0.05]) ;
```

```matlab
effect = stat_model.t; 
effect(pval.P>0.01) = 0; 
SurfStatViewData(effect, surf_white, 'RFT');
colormap([0.8 .8 .8; ice/255])
SurfStatColLim([0 0.05]) ;
```
