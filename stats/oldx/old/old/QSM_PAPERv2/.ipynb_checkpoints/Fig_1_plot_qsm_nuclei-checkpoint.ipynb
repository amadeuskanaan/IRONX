```python
%matplotlib inline
import os 
import numpy as np
import nibabel as nb 
from numpy import mean, std
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import colors
from matplotlib.colors import ListedColormap
#from nilearn.plotting import plot_anat
plt.rcParams['font.family'] = 'Myriad Pro'
sns.set_style('white')
from scipy import ndimage

datadir = '/Users/kanaan/Google Drive/TS-EUROTRAIN/RESULTS_QSM/NUCLEI_FEB17'
savefigdir    = '/Users/kanaan/Google Drive/TS-EUROTRAIN/Papers/2016_QSM_paper/Figures_python'

palette = sns.color_palette("hls", 20)
sns.palplot(palette)

alpha = 0.7
```

```python
def edge_detect_ero(i):
    ero  = ndimage.binary_erosion(i,iterations=1).astype(i.dtype)
    edge = i -ero
    edge[edge==0]=np.nan
    return edge

def edge_detect_first(i):
    ero =  ndimage.binary_erosion(i, iterations=2).astype(i.dtype)
    dil =  ndimage.binary_dilation(i, iterations=1).astype(i.dtype)
    sep = dil-ero
    sep = ndimage.binary_erosion(sep, iterations=1).astype(sep.dtype)
    sep[sep==0]=np.nan
    return sep

def get_first_data(subject, nucleus):
        return np.rot90(nb.load(os.path.join(datadir, subject, 'FIRST_HYBRID-L_%s_first_thr.nii.gz'%nucleus)).get_data()) \
                  + np.rot90(nb.load(os.path.join(datadir, subject, 'FIRST_HYBRID-R_%s_first_thr.nii.gz'%nucleus)).get_data()) 

def get_atak_data(subject, nucleus):
        return np.rot90(nb.load(os.path.join(datadir, subject, 'L_%s.nii.gz'%nucleus)).get_data()) \
                  + np.rot90(nb.load(os.path.join(datadir, subject, 'R_%s.nii.gz'%nucleus)).get_data()) 

def plot_qsm(subject,cut, scale = [-0.2,0.2], cmapx  = 'r', alpha_mask = 0.35, flip = 0):
    if subject == 'mni':
        qsm = np.rot90(nb.load(os.path.join(datadir[:-12], 'ATAK/mean.nii.gz')).get_data()) 
    else:
        qsm = np.rot90(nb.load(os.path.join(datadir, subject, 'QSM_denoised.nii.gz')).get_data())
    #uni = np.rot90(nb.load(os.path.join(datadir, subject,, 'MP2RAGE2FLASH_BRAIN.nii.gz')).get_data()) #/  350
    
    if flip:
        qsm = qsm * -1
    fig = plt.figure()
    fig.set_size_inches(10, 5)
    fig.subplots_adjust(wspace=0.05, hspace= 0.05)
    cmap = colors.ListedColormap(cmapx)
    intp_mask = 'mitchell'
    
    ax1 = plt.subplot2grid((3,2), (0,0), rowspan=3)
    ax1.set_xlim(15, 190)
    ax1.set_ylim(255,5)
    ax1.imshow(qsm[:,:,cut[0]], vmin = scale[0], vmax = scale[1])
    for nuc in ['Caud','Puta','Pall', 'Thal']:
        mask = get_first_data(subject, nuc)
        #ax1.imshow(edge_detect_first(mask)[:,:,cut[0]], cmap = cmap,  interpolation=intp_mask)
        mask[mask==0]=np.nan
        ax1.imshow(mask[:,:,cut[0]], cmap = cmap,alpha =alpha_mask, interpolation=intp_mask)
    
    ax2 = plt.subplot2grid((3,2), (0,1), rowspan=1)
    ax2.imshow(qsm[:,:,cut[1]], vmin = scale[0], vmax = scale[1])
    ax2.set_xlim(45, 155)
    ax2.set_ylim(160, 110)
    for nuc in ['RN','SN']:
        mask = get_atak_data(subject, nuc)
        #ax2.imshow(edge_detect_ero(mask)[:,:,cut[1]], cmap = cmap,  interpolation=intp_mask)
        mask[mask==0]=np.nan
        ax2.imshow(mask[:,:,cut[1]], cmap = cmap,alpha =alpha_mask, interpolation=intp_mask)
    
    ax3 = plt.subplot2grid((3,2), (1,1), rowspan=1)
    ax3.imshow(np.rot90(qsm[cut[2],:,:]), vmin = scale[0], vmax = scale[1])
    ax3.set_xlim(45, 155)
    ax3.set_ylim(125, 75)
    for nuc in ['STN','SN']:
        mask = get_atak_data(subject, nuc)
        #ax3.imshow(np.rot90(edge_detect_ero(mask)[cut[2],:,:]), cmap = cmap,  interpolation=intp_mask)
        mask[mask==0]=np.nan
        ax3.imshow(np.rot90(mask[cut[2],:,:]), cmap = cmap, alpha =alpha_mask, interpolation=intp_mask)
          
    ax4 = plt.subplot2grid((3,2), (2,1), rowspan=1)
    ax4.imshow(qsm[:,:,cut[3]], vmin = scale[0], vmax = scale[1])
    ax4.set_xlim(50, 160)
    ax4.set_ylim(210, 160)
    for nuc in ['DN']:
        mask = get_atak_data(subject, nuc)
        #ax4.imshow(edge_detect_ero(mask)[:,:,cut[3]], cmap = cmap,  interpolation=intp_mask)
        mask[mask==0]=np.nan
        ax4.imshow(mask[:,:,cut[3]], cmap = cmap,alpha =alpha_mask,  interpolation=intp_mask)
        
    for ax in [ax1,ax2,ax3,ax4]:
        plt.setp( ax.get_xticklabels(), visible=False)
        plt.setp( ax.get_yticklabels(), visible=False)

    
    
    
    
    
    
    
    
    
    
    
```
