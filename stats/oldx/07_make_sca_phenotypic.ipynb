```python
def get_df(popname):
    df = pd.read_csv(os.path.join(pheno_dir, 'df_%s_qc.csv'%popname),index_col = 0)
    return df 

controls = get_df('controls')
patients = get_df('patients')

#controls['Group'] = 0
#patients['Group'] = 1

pheno2 = '/Users/kanaan/SCR/workspace/project_touretome/phenotypic/df_dcm/dicomhdr_leipzig.csv'
pheno2 = pd.read_csv(pheno2, index_col = 1)

name_dict = dict(pheno2['Unnamed: 0'])
```

```python
name_dict['SMVX']
```

```python
cols = ['Age', 'Gender', 'Group', 'QI1_MAG', 'EFC_MAG', 'ACC_FD']

df = pd.concat([controls, patients])
df = df.drop([i for i in df.columns if i not in cols], axis = 1)
df.index.name='idx'

#df = df.dropna()
df = df.drop(['FA2T','GHAT', 'BH5T', 'LMIT', 'RA7T','RMJP']) 
fd_mu = np.mean(df.ACC_FD)
df = df.fillna(fd_mu)

df['idx2'] = 'X'
for i in df.index:
    df.loc[i,'idx2'] = name_dict[i]

df = df.set_index('idx2')
df 
```
