# Clinical data decomposition using PCA 

```r
library(psych)

setwd('/Users/kanaaax/Google Drive/TS-EUROTRAIN/Papers/2016_QSM_paper/Figures_python_v2')

datadir = '/Users/kanaaax/Google Drive/TS-EUROTRAIN/RESULTS_QSMv3/dataframes/phenotypic'
dfp = read.table(file.path(datadir,'df_patients_qc.csv'), sep = ",", header=TRUE, row.names = 1)

data  = dfp[,c('CLN_YGTSS_Motoric_Score',
               'CLN_YGTSS_Vocal_Score', 
               'CLN_RVTRS',
               'CLN_puts',
               'CLN_OCIR_total_score',
               'CLN_DSM4_ADHD_Score',
               'CLN_CAARS_Score_ADHS_Symptoms_Total_Cat_G_T_Score',
               'CLN_BAI',
               'CLN_BDI12',
               'CLN_MADRS'
               )] 
               
#write.csv(data, file = file.path(datadir,'cln.csv'))
```

```r
data = scale(data)

colnames(data) = c('YGTSS-Motor', 
                   'YGTSS-Vocal', 
                   'RVTRS', 
                   'PUTS', 
                   'OCI-R', 
                   'DSM4-ADHD',
                   'CAARS', #'YBOCS', 
                   'BAI', 
                   'BDI-II', 
                   'MADRS')
```

```r
# Perform PCA
PCA = principal(data, missing=TRUE,impute="median", rotate="varimax",nfactors=4)
PCA
```

```r
library(corrplot)
library(psych)
corrplot(PCA$loadings, method="ellipse")
```

```r
ggbiplot(PCA, obs.scale = 0, var.scale = 0, ellipse = FALSE, circle = TRUE)
```
