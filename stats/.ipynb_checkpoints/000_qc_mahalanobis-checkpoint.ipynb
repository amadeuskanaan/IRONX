```r
library(MVN)

datadir = '/Users/kanaaax/Google Drive/TS-EUROTRAIN/RESULTS_QSMv3/OLD/AUG5/phenotypic'
dfc = read.table(file.path(datadir,'df_raw_controls.csv'), sep = ",", header=TRUE, row.names = 1)
dfp = read.table(file.path(datadir,'df_raw_patients.csv'), sep = ",", header=TRUE, row.names = 1)

df = rbind(dfc, dfp)
notneededdata = c()
data  = df[,c('EFC_MAG','QI1_MAG')] 

result =mvn(data, subset = NULL, scale = FALSE, desc = TRUE, transform = "none", R = 1000,
            multivariatePlot = "scatter", multivariateOutlierMethod = "quan", bc = FALSE, bcType = "rounded", 
            showOutliers = FALSE, showNewData = FALSE)
```
